```python
from matplotlib import pyplot
import textwrap

#sns.set_theme(style="whitegrid")
sns.set_style('white')
pyplot.figure(figsize=(24, 4))
sns.set(font_scale=1.5)

ax = sns.barplot(x="Scenario", y="Average Goal Difference", hue='Evaluation Script', data=df)

from matplotlib.ticker import MultipleLocator
ax.yaxis.set_major_locator(MultipleLocator(0.2))
ax.set(ylim=(-1, 1))
ax.set_xticklabels(['\n'.join(textwrap.wrap(ticklabel.get_text(),8)) for ticklabel in ax.get_xticklabels()])
ax.set(xlabel=None)
pyplot.legend(loc='upper right')
ax.figure.savefig("testing.png", bbox_inches="tight")
```

```python
from matplotlib import pyplot
import textwrap

#sns.set_theme(style="whitegrid")
sns.set_style('white')
pyplot.figure(figsize=(24, 4))
sns.set(font_scale=1.5)

ax = sns.barplot(x="Scenario", y="Average Goal Difference", hue='Evaluation Script', data=df[df["Scenario Type"]=="Defense"])

from matplotlib.ticker import MultipleLocator
ax.yaxis.set_major_locator(MultipleLocator(0.2))
ax.set(ylim=(-0.8, 0))
ax.set_xticklabels(['\n'.join(textwrap.wrap(ticklabel.get_text(),8)) for ticklabel in ax.get_xticklabels()])
ax.set(xlabel=None)
pyplot.legend(loc='lower right')
ax.figure.savefig("testing_defense.png", bbox_inches="tight")
```

```python
from matplotlib import pyplot
import textwrap

#sns.set_theme(style="whitegrid")
sns.set_style('white')
pyplot.figure(figsize=(24, 4))
sns.set(font_scale=1.5)

ax = sns.barplot(x="Scenario", y="Average Goal Difference", hue='Evaluation Script', data=df[df["Scenario Type"]!="Defense"])

from matplotlib.ticker import MultipleLocator
ax.yaxis.set_major_locator(MultipleLocator(0.2))
ax.set(ylim=(0, 1))
ax.set_xticklabels(['\n'.join(textwrap.wrap(ticklabel.get_text(),8)) for ticklabel in ax.get_xticklabels()])
ax.set(xlabel=None)
pyplot.legend(loc='upper right')
ax.figure.savefig("testing_offense.png", bbox_inches="tight")
```
